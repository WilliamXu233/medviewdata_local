```python
# Import MongoDB client and server API interface
from pymongo import MongoClient
import re
# Ping the database to verify the connection
from typing import List, Dict, Any, Set, Tuple
from collections import Counter
import math

def connect_to_mongodb(connection_string: str, db_name: str, collection_name: str):
    """Connect to MongoDB and return the specified collection."""
# Create a new MongoDB client instance using the URI
    client = MongoClient(connection_string)
# Connect to database and select collection
    db = client[db_name]
    collection = db[collection_name]
    return collection

def calculate_cosine_similarity(input_question: str, stored_question: str) -> float:
    """
    Calculate cosine similarity between two questions.
    
    Cosine similarity = dot product of vectors / (magnitude of vector1 * magnitude of vector2)
    """
    # Normalize questions (lowercase, remove punctuation)
    def normalize(text: str) -> str:
        return re.sub(r'[^\w\s]', '', text.lower())
    
    input_normalized = normalize(input_question)
    stored_normalized = normalize(stored_question)
    
    # Split into words
    input_words = input_normalized.split()
    stored_words = stored_normalized.split()
    
    # Create word frequency vectors
    input_vector = Counter(input_words)
    stored_vector = Counter(stored_words)
    
    # Find all unique words
    all_words = set(input_vector.keys()).union(set(stored_vector.keys()))
    
    # Calculate dot product
    dot_product = sum(input_vector.get(word, 0) * stored_vector.get(word, 0) for word in all_words)
    
    # Calculate magnitudes
    input_magnitude = math.sqrt(sum(input_vector.get(word, 0) ** 2 for word in all_words))
    stored_magnitude = math.sqrt(sum(stored_vector.get(word, 0) ** 2 for word in all_words))
    
    # Avoid division by zero
    if input_magnitude == 0 or stored_magnitude == 0:
        return 0.0
    
    # Calculate cosine similarity
    return dot_product / (input_magnitude * stored_magnitude)

def find_similar_question(collection, input_question: str, min_similarity: float = 0.3) -> Dict[str, Any]:
    """
    Find the most similar question in the MongoDB collection and return its answer.
    
    Args:
        collection: MongoDB collection
        input_question: The question to match
        min_similarity: Minimum similarity threshold (0-1)
        
    Returns:
        Dictionary containing the matched question, answer, and similarity score
    """
    all_documents = collection.find({})
    best_match = None
    highest_similarity = min_similarity
    
    for doc in all_documents:
        model = doc.get("model", "Unknown")
        questions = doc.get("questions", [])
        
        for q_item in questions:
            stored_question = q_item.get("question", "")
            answer = q_item.get("answer", "")
            
            similarity = calculate_cosine_similarity(input_question, stored_question)
            
            if similarity > highest_similarity:
                highest_similarity = similarity
                best_match = {
                    "model": model,
                    "matched_question": stored_question,
                    "answer": answer,
                    "similarity_score": similarity
                }
    
    return best_match

def main():
    # Example usage
    connection_string = "mongodb+srv://<username>:<password>@harris2004wang:zFvz9bMbT7scRJG4@mdvpre.5hiqsmm.mongodb.net/?appName=mdvpre"
    db_name = "mdvd"
    collection_name = "mdve"
    
    # Connect to MongoDB
    try:
        collection = connect_to_mongodb(connection_string, db_name, collection_name)
        
        # Example question to match
        input_question = "How do I clean my Lowenstein CPAP machine?"
        
        # Find similar question and answer
        result = find_similar_question(collection, input_question)
        
        if result:
            print(f"Model: {result['model']}")
            print(f"Matched Question: {result['matched_question']}")
            print(f"Similarity Score: {result['similarity_score']:.2f}")
            print("\nAnswer:")
            print(result['answer'])
        else:
            print("No similar question found.")
            
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
